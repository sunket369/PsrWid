```python
def generate_pulse(length, amp, mean, std, noise_mean, noise_std, phase_shift=0, noise=True):
    """
    Generate a Gaussian pulse profile with noise.
    amp, length, mean, std - of the Gaussian pulse
    noise_mean, noise_std  - of the Gaussian noise
    phase_shift            - to shift the pulse
    noise                  - to add noise or not
    

    """
    phi = np.linspace(0, 2*np.pi, length)
    pulse = amp * np.exp(-((phi - mean)**2) / (2*std**2))
    if noise:
        noise = np.random.normal(noise_mean, noise_std, length)
        pulse += noise
    return np.roll(pulse, phase_shift)

amp = 1
mean = np.pi
std = 0.1
noise_mean = 0
noise_std = 0.5
length = 1024
phase_shift = 0

I = generate_pulse(length, amp, mean, std, noise_mean, noise_std, phase_shift=0, noise=True)
analysis = PsrWid(I)
analysis.plot_results(get_error=True, xlim=(0, 2*np.pi))
```
